```python
#Import pandas
import pandas as pd
```

```python
#read CSV file
netflixdata=pd.read_csv("Netflix.csv")
```

```python
#create dataframe
df=pd.DataFrame(netflixdata)
```

```python
#Data preview 
df.head()
```

```python
#Deleting unnecessary columns
columns_to_drop=["Title","Tags","Hidden Gem Score","Country Availability","Director","Writer","Actors","Rotten Tomatoes Score","View Rating","Metacritic Score","Awards Nominated For","Boxoffice","Netflix Release Date","Production House","Netflix Link","IMDb Link","Summary","IMDb Votes","Image","Poster","TMDb Trailer","Trailer Site","Languages","Genre","Runtime"]
df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
#Data preview 
df.head()
```

```python
#Null value
null_counts = df.isnull().sum()
null_dict = null_counts.to_dict()
null_dict
```

```python
#None Null values
non_null_counts = df.notnull().sum()
non_null_dict = non_null_counts.to_dict()
non_null_dict
```

```python
#checking whats the percentage of null value out of my data
percentage_of_null={key: (null_dict[key] / non_null_dict[key])*100 for key in null_dict}
percentage_of_null
```

```python
#I am removing all rows with release date that is null.
df.dropna(subset=['Release Date'], how="any", inplace=True)
```

```python
#I am removing all rows with IMDb Score that is null.
df.dropna(subset=['IMDb Score'], how="any", inplace=True)
```

```python
#replace Awards Received nulls by "Not Awarded"
df['Awards Received'].fillna("Not Awarded", inplace=True)
```

```python
#change the format type of column "Release Date" to dates format
df["Release Date"]=pd.to_datetime(df["Release Date"])
```

```python
df
```

```python
#splitting the release dates into 20 years period
years_per_category = 20
min_year = 1910
max_year = 2022
num_categories = (max_year - min_year) // years_per_category + 1
category_bins = [(min_year + i * years_per_category, min_year + (i + 1) * years_per_category - 1) for i in range(num_categories)]
labels = [f'{start}-{end}' for start, end in category_bins]
df['Release Period'] = pd.cut(df['Release Date'].dt.year, bins=[start for start, end in category_bins] + [max_year + 1], labels=labels, right=False)
```

```python
df.to_csv('output.csv', index=True)
```
